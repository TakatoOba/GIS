```python
gdf = gpd.read_file('B002005212020DDSWC08220.zip')
```

```python
gdf = gdf.dissolve(by='S_NAME',aggfunc='sum').reset_index()
```

```python
fig,ax = plt.subplots()
gdf.plot(ax=ax)
```

```python
fig,ax = plt.subplots()
gdf.plot(ax=ax,
         column='JINKO',
         legend=True);
```

```python
fig,ax = plt.subplots(figsize=(10,10))
gdf.plot(ax=ax,
         column='JINKO',
         legend=True);
```

```python
fig,ax = plt.subplots(figsize=(10,10))
gdf.plot(ax=ax,
         column='JINKO',
         legend=True,
         cmap='Reds');
```
